### Stopwords in NLTK

### üìò 1. Definition: What Are Stopwords?

Stopwords are the most frequently occurring function words in a language that do not carry significant meaning and are often removed during text preprocessing in NLP.

Examples:
["the", "is", "and", "a", "an", "in", "on", "at", "of", "for", "to"]

### üß† 2. Purpose in NLP
Reduces dimensionality of the dataset

Improves model efficiency and training speed

Focuses on semantic content (keywords) rather than grammatical structure



### üîß Example: English Stopword Removal

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')

text = "This is a sample sentence, showing off the stop words filtration."
stop_words = set(stopwords.words('english'))

words = word_tokenize(text.lower())
filtered = [word for word in words if word not in stop_words and word.isalnum()]
print("Filtered Text:", filtered)
```

### üìä 4. Advantages of Removing Stopwords
| Benefit                      | Description                                         |
| ---------------------------- | --------------------------------------------------- |
| ‚úÖ **Reduces Noise**          | Removes common, non-informative words               |
| ‚úÖ **Improves Focus**         | Helps models focus on meaningful tokens             |
| ‚úÖ **Decreases Feature Size** | Reduces vocabulary and increases sparsity           |
| ‚úÖ **Faster Computation**     | Fewer tokens to process means faster model training |


### ‚ö†Ô∏è 5. Disadvantages / Risks
| Drawback                | Description                                                                                          |
| ----------------------- | ---------------------------------------------------------------------------------------------------- |
| ‚ùå **Context Loss**      | Important context may be lost if stopwords are removed indiscriminately (e.g., ‚ÄúTo be or not to be‚Äù) |
| ‚ùå **Domain Dependency** | Some stopwords may be relevant in specific domains (e.g., ‚Äúnot‚Äù in sentiment analysis)               |
| ‚ùå **Hardcoded Lists**   | Static lists may not generalize well to specific corpora or dialects                                 |
| ‚ùå **Over-removal**      | Risk of removing words that are semantically meaningful in a given context                           |

### üõ†Ô∏è 6. Best Practices
| Practice                                                                                          | Recommendation |
| ------------------------------------------------------------------------------------------------- | -------------- |
| ‚úÖ Use **customized stopword lists** for domain-specific NLP                                       |                |
| ‚úÖ Keep **negation words** (`no`, `not`, `never`) for sentiment tasks                              |                |
| ‚úÖ Combine NLTK with **TF-IDF**, so that frequent non-informative terms are naturally downweighted |                |
| ‚úÖ Consider **contextual models** (BERT, spaCy) that don‚Äôt need stopword removal at all            |                |

```python
## Speech Of DR APJ Abdul Kalam
paragraph = """I have three visions for India. In 3000 years of our history, people from all over 
               the world have come and invaded us, captured our lands, conquered our minds. 
               From Alexander onwards, the Greeks, the Turks, the Moguls, the Portuguese, the British,
               the French, the Dutch, all of them came and looted us, took over what was ours. 
               Yet we have not done this to any other nation. We have not conquered anyone. 
               We have not grabbed their land, their culture, 
               their history and tried to enforce our way of life on them. 
               Why? Because we respect the freedom of others.That is why my 
               first vision is that of freedom. I believe that India got its first vision of 
               this in 1857, when we started the War of Independence. It is this freedom that
               we must protect and nurture and build on. If we are not free, no one will respect us.
               My second vision for India‚Äôs development. For fifty years we have been a developing nation.
               It is time we see ourselves as a developed nation. We are among the top 5 nations of the world
               in terms of GDP. We have a 10 percent growth rate in most areas. Our poverty levels are falling.
               Our achievements are being globally recognised today. Yet we lack the self-confidence to
               see ourselves as a developed nation, self-reliant and self-assured. Isn‚Äôt this incorrect?
               I have a third vision. India must stand up to the world. Because I believe that unless India 
               stands up to the world, no one will respect us. Only strength respects strength. We must be 
               strong not only as a military power but also as an economic power. Both must go hand-in-hand. 
               My good fortune was to have worked with three great minds. Dr. Vikram Sarabhai of the Dept. of 
               space, Professor Satish Dhawan, who succeeded him and Dr. Brahm Prakash, father of nuclear material.
               I was lucky to have worked with all three of them closely and consider this the great opportunity of my life. 
               I see four milestones in my career"""
```

```python
from nltk.stem import PorterStemmer, SnowballStemmer, WordNetLemmatizer
```

```python
para_stopwords : list =[]
para_actual_words : list =[]
```

```python
sentenses = nltk.sent_tokenize(paragraph)
```

```python
sentenses
```

```python
stemmer = PorterStemmer()
for sentense in sentenses:
    words = nltk.word_tokenize(sentense)
    for word in words:
        if stemmer.stem(word) not in stopwords.words('english'):
            para_actual_words.append(word)
        else:
            para_stopwords.append(word)

' '.join(para_actual_words)  # Join stopwords into a single string
```

```python

stemmer1 = SnowballStemmer('english')
for sentense in sentenses:
    words = nltk.word_tokenize(sentense)
    for word in words:
        if stemmer1.stem(word) not in stopwords.words('english'):
            para_actual_words.append(word)
        else:
            para_stopwords.append(word)

' '.join(para_actual_words)  # Join stopwords into a single string
```

```python
lemmatize = WordNetLemmatizer()
for sentense in sentenses:
    words = nltk.word_tokenize(sentense)
    for word in words:
        if lemmatize.lemmatize(word.lower(),pos='v') not in stopwords.words('english'):
            para_actual_words.append(word)
        else:
            para_stopwords.append(word)

' '.join(para_actual_words)  # Join stopwords into a single string
```
