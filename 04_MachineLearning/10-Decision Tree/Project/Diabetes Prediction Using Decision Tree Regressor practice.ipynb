```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
## find target variable
# The diabetes dataset does not have 'target_names', only 'target'
# To see what the target represents, check the description:
print(diabetes['DESCR'].split('Target:')[1].split('\n')[0].strip())
```

```python
## 
X = pd.DataFrame(diabetes['data'], columns=diabetes['feature_names'])
y = pd.Series(diabetes['target'], name='diabetes_progression')
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
reggessor = DecisionTreeRegressor()
```

```python
X_train.corr()
```

```python
import seaborn as sns
plt.Figure(figsize=(20,20))
sns.heatmap(X_train.corr(),annot=True)
```

```python
reggessor.fit(X_train,y_train)
```

```python
y_pred = reggessor.predict(X_test)
```

```python
## Hyperparameter Tuning
from sklearn.model_selection import GridSearchCV
param={
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}
```

```python
regressor=DecisionTreeRegressor()
```

```python
grid=GridSearchCV(regressor,param_grid=param,cv=5,scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```
