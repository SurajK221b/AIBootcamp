### Skewness is a statistical measure that quantifies the asymmetry of a probability distribution around its mean. It plays a critical role in exploratory data analysis (EDA), feature engineering, and model diagnostics.

#### ðŸ“Š Types of Skewness
| **Skewness Type**              | **Direction**     | **Characteristics**                | **Mean-Median Mode Relationship** |
| ------------------------------ | ----------------- | ---------------------------------- | --------------------------------- |
| **Symmetrical (Zero Skew)**    | Balanced          | Bell-shaped, Normal Distribution   | Mean = Median = Mode              |
| **Positive Skew (Right Skew)** | Tail to the right | Long right tail, few high outliers | Mean > Median > Mode              |
| **Negative Skew (Left Skew)**  | Tail to the left  | Long left tail, few low outliers   | Mean < Median < Mode              |


### ðŸ“˜ Data Science Implications of Skewness
| **Use Case**                      | **Skewness Insight**                    | **Recommended Action**                       |
| --------------------------------- | --------------------------------------- | -------------------------------------------- |
| **Feature Distribution Analysis** | Detect asymmetry                        | Apply transformations (log, sqrt, Box-Cox)   |
| **Outlier Detection**             | Skewness often driven by outliers       | Handle via capping or imputation             |
| **Model Assumptions**             | Many models assume normality            | Address skew before regression or clustering |
| **EDA and Visualization**         | Skewness complements histogram insights | Use alongside mean/median indicators         |

###ðŸ“ˆ Real-World Examples

| **Dataset Attribute** | **Likely Skew Type** | **Rationale**                                |
| --------------------- | -------------------- | -------------------------------------------- |
| **Income**            | Positive             | Few people earn much more than average       |
| **Loan Default Time** | Negative             | Most people default early or not at all      |
| **Customer Age**      | Often Symmetric      | Normally distributed in balanced populations |


### ðŸŽ¯ Interview-Style Q&A
| **Question**                                        | **Answer**                                                                        |
| --------------------------------------------------- | --------------------------------------------------------------------------------- |
| What does positive skew indicate in a dataset?      | Most values are concentrated on the left; a few large values pull the tail right. |
| How do you address skewed features before modeling? | Apply log, square root, or Box-Cox transformations.                               |
| Why is skewness important in regression analysis?   | Many regression models assume that residuals are normally distributed.            |
| What does it mean if skewness is close to zero?     | The distribution is approximately symmetrical.                                    |


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('kohli_scores.xlsx')
df.head()
```

```python
sns.histplot(df.score, kde=True)
```

```python
mean = df.score.mean()
std = df.score.std()

mean, std
```

```python
sns.histplot(df.score, kde=True)
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.axvline(mean + std, color='green', linestyle='--', label='Mean + 1 Std Dev')
plt.axvline(mean - std, color='blue', linestyle='--', label='Mean - 1 Std Dev')
plt.legend()
```
