```python
pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm'])
```

```python
X= pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm'])
y= iris['target']
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_text,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train
```

```python
X_text
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
treeClass = DecisionTreeClassifier(max_depth=2)
```

```python
treeClass.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(20, 10))
tree.plot_tree(treeClass, filled=True)
```

```python
y_pred = treeClass.predict(X_text)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm=confusion_matrix(y_test,y_pred)
cr=classification_report(y_test,y_pred)
cm,cr
```

```python
print(cm)
print(cr)
```

```python
param={
    'criterion':['gini','entropy', 'log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
param
```
