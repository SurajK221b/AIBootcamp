### üéØ Principal Component Analysis (PCA): Executive Summary
| Feature       | Description                                                                       |
| ------------- | --------------------------------------------------------------------------------- |
| **Category**  | Unsupervised Learning, Dimensionality Reduction                                   |
| **Objective** | Reduce the number of features while preserving the maximum variance (information) |
| **Type**      | Linear transformation                                                             |
| **Output**    | Principal Components (uncorrelated, ranked by variance explained)                 |


### üß† Core Intuition
PCA identifies the axes (principal components) along which the data varies the most. It rotates the original coordinate system to align with the directions of maximum variance and projects the data onto a smaller subspace without losing significant information.

### üî¨ When to Use PCA
High-dimensional data with potential multicollinearity.

Need to visualize data in 2D or 3D.

As a preprocessing step before applying clustering or classification models.

When feature selection is not practical due to redundancy or correlation.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Load dataset
data = load_iris()
X = data.data
y = data.target

# Standardize
X_scaled = StandardScaler().fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Explained variance
print("Explained Variance Ratio:", pca.explained_variance_ratio_)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA - Iris Dataset")
plt.grid(True)
plt.show()
```
